```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv('ex1data1.csv')
```

```python
data
```

```python
data1=data.Population
data2=data.Profit
```

```python
plt.scatter(data1,data2,c='r')
plt.title("Population-Food Profit")
plt.xlabel("Population in 10,000")
plt.ylabel("Profit in $10,000")
```

```python
def linear_regression(x,y,theta_zero=0,theta_one=0,iter=1000,alpha=0.02):
  m=int(len(y))
  for i in range(iter):
    hx=theta_zero+theta_one*x
    cost=(1/(2*m))*sum([d**2 for d in (hx-y)])
    gradient_zero=1/m*(sum(hx-y))
    gradient_one=1/m*(sum((hx-y)*x))
    theta_zero=theta_zero-(alpha*gradient_zero)
    theta_one=theta_one-(alpha*gradient_one)
    print("theta_zero=",theta_zero,"theta_one=",theta_one,"Cost=",cost,"iteration=",i)
  return theta_zero, theta_one,cost
theta_zero,theta_one,cost=linear_regression(data1,data2)
```

```python
line=[(theta_zero+theta_one*n) for n in data1]
```

```python
plt.scatter(data1,data2,c='r',label='training data')
plt.plot(data1,line,label='Linear Regression')
plt.title("Population-Food Profit")
plt.xlabel("Population in 10,000")
plt.ylabel("Profit in $10,000")
plt.legend()
```
